```python
import json
import uuid
import numpy as np
from tqdm import tqdm
from qdrant_client import QdrantClient, models
from FlagEmbedding import BGEM3FlagModel

model = BGEM3FlagModel('BAAI/bge-m3', use_fp16=True)


db_path = "/kaggle/working/medical_qdrant_db"
client = QdrantClient(path=db_path) 
```

```python
collection_name = "arabic_medical_HybridRAG"

client.create_collection(
    collection_name=collection_name,
    vectors_config={
        "dense": models.VectorParams(
            size=1024, 
            distance=models.Distance.COSINE
        )
    },
    sparse_vectors_config={
        "sparse": models.SparseVectorParams(
            modifier=models.Modifier.IDF
        )
    }
)
```

```python
file_path = "/kaggle/input/datasets/shamsccs/medrag/RAG_cleaned_overlapped.jsonl"
dataset = []

with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        dataset.append(json.loads(line))

BATCH_SIZE = 16 

for i in tqdm(range(0, len(dataset), BATCH_SIZE)):
    batch = dataset[i:i + BATCH_SIZE]
    
    texts = [doc["content"] for doc in batch]
    
    payloads = [
        {
            "chunk_id": doc["chunk_id"], 
            "text": doc["content"], 
            "source": doc.get("source", ""),
            "page_number": doc.get("page_number", "")
        } 
        for doc in batch
    ]
    
    ids = [str(uuid.UUID(doc["chunk_id"])) for doc in batch]
    
    embeddings = model.encode(
        texts, 
        max_length=1024,
        return_dense=True, 
        return_sparse=True, 
        return_colbert_vecs=False, 
        batch_size=BATCH_SIZE
    )
    
    dense_vecs = embeddings['dense_vecs'].tolist()
    
    sparse_vecs = []
    for weight_dict in embeddings['lexical_weights']:
        indices = [int(k) for k in weight_dict.keys()]
        values = list(weight_dict.values())
        sparse_vecs.append(models.SparseVector(indices=indices, values=values))
        
    points = [
        models.PointStruct(
            id=ids[idx],
            vector={
                "dense": dense_vecs[idx],
                "sparse": sparse_vecs[idx]
            },
            payload=payloads[idx]
        )
        for idx in range(len(texts))
    ]
    
    client.upsert(collection_name=collection_name, points=points)
```

```python
query = "نسبة السكر في الدم"
print(query)

query_embeddings = model.encode(
    [query], 
    return_dense=True, 
    return_sparse=True, 
    return_colbert_vecs=False
)

query_dense = query_embeddings['dense_vecs'][0].tolist()
query_weight_dict = query_embeddings['lexical_weights'][0]

query_sparse = models.SparseVector(
    indices=[int(k) for k in query_weight_dict.keys()],
    values=list(query_weight_dict.values())
)

search_result = client.query_points(
    collection_name=collection_name,
    prefetch=[
        models.Prefetch(query=query_dense, using="dense", limit=10),
        models.Prefetch(query=query_sparse, using="sparse", limit=10),
    ],
    query=models.FusionQuery(fusion=models.Fusion.RRF),
    limit=3
)

for result in search_result.points:
    print(f"Score: {result.score:.4f}")
    print(f"Qdrant Point ID: {result.id}")
    print(f"Source: {result.payload.get('source', '')} | Page: {result.payload.get('page_number', '')}")
    print(f"Text: {result.payload['text'][:350]}...\n")
    print("-" * 50)
```
